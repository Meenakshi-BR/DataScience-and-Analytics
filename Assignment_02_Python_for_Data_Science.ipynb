5.Perform meaningful visualizations using the dataset.Bring atleast 3 meaningful visualizations.

Boxplot

```python
df1=df.set_index('Classification')
plt.boxplot(df1)
plt.show()
```

Scatter Plot

```python
sns.scatterplot(x='SL',y='SW',data=df,hue='Classification',palette='deep')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

PairPlot

```python
sns.pairplot(df,hue='Classification')
plt.suptitle('Pairplot of Iris dataset',y=1.02,fontsize=16)
plt.show()
```

HeatMap

```python
sns.heatmap(df.select_dtypes(include='number').corr())
plt.title('Correlation Heatmap of Iris features')
plt.show()
```
